```python
import pandas as pd
import glob
import os
```

```python
csv_files = ['wholesale_price.csv','oil_price.csv','inr_usd.csv','iip_index.csv','cpi_index.csv']
```

```python
df_list=[]
for file in csv_files:
  df_list.append(pd.read_csv(file))
```

```python
combined_df = pd.concat(df_list, axis=1)
print("Combined data shape:",combined_df.shape)
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(combined_df)
```

```python
combined_df[['year','month']] = combined_df['Date'].str.split('-', expand=True)
combined_df.head()
```

```python
combined_df = combined_df.drop(columns=['Date'])
```

arrange the columns


```python
remain_columns = ['year','month','YoY_Inflation_Rate']
other_columns = [col for col in combined_df.columns if col not in remain_columns]
arrange = ['year','month'] + other_columns + ['YoY_Inflation_Rate']
combined_df = combined_df[arrange]
```

```python
combined_df.head()
```

cleaning the data's

```python
print(combined_df.dtypes)
```

```python
print(combined_df.isnull().sum())
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(combined_df)
```

change the data type and remove the nan values

```python
#droping the nan values
combined_df = combined_df.dropna()
```

```python
combined_df['year'] = combined_df['year'].astype(int)
combined_df['month'] = combined_df['month'].astype(int)
```

```python
print(combined_df.dtypes)
```

```python
print(combined_df.isnull().sum())
```

now ill export the proper csv in next colab file for proper feature scalling and train

```python
# Saving the cleaned DF to a new CSV file
combined_df.to_csv('inflation_detection.csv', index=False)

print("File saved successfully!")
```
